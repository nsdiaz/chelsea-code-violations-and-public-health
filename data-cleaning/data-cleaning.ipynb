### Creating a new variable to incorporate the target area

```python
# Read the dataset
target_area = pd.read_csv('target_area.csv', skiprows=6)

# Create new column with a default value of 1
target_area['Target Area'] = 1

# Select only the columns that we care about
target_area = target_area[['Parcel ID', 'Target Area']]

# Merge it with the rest of the dataset
chelsea = chelsea.merge(target_area, on = 'Parcel ID', how = 'left')

# Fill the NaNs in the target area column
chelsea['Target Area'].fillna(0, inplace = True)
```

```python
chelsea.shape
```

```python
chelsea['any_coh'] = chelsea['Certificate of Habitability by Status']
chelsea.loc[chelsea.any_coh >= 1, 'any_coh'] = 1
chelsea.any_coh.fillna(0, inplace = True)
```

```python
# Checking how many COH parcels are outside target area
chelsea['test parcels'] = np.where(((chelsea.any_coh == 1) & (chelsea['Target Area'] == 0)), 1, 0)
print(chelsea['test parcels'].sum())
```

```python
chelsea.shape
```

### New variable with  recent police incidents

```python
recent_police = pd.read_csv('police_after2015.csv', skiprows=4)
recent_police.columns = ['Parcel ID', 'Police Incidents After 2015']
recent_police['Police Incidents After 2015'].value_counts()
recent_police.fillna(0, inplace = True)
```

```python
chelsea = chelsea.merge(recent_police, on = 'Parcel ID', how = 'left')
```

```python
chelsea.shape
```

```python
# Filling all the violation columns
chelsea.fillna(0, inplace = True)
```

### Dealing with duplicates

```python
# Halving duplicate violations
print(chelsea['Certificate of Habitability Violations - Hot Water'].sum())
chelsea['Certificate of Habitability Violations - Hot Water'] = chelsea['Certificate of Habitability Violations - Hot Water']/2
print(chelsea['Certificate of Habitability Violations - Hot Water'].sum())

print(chelsea['Certificate of Habitability Violations - Other Violation'].sum())
chelsea['Certificate of Habitability Violations - Other Violation'] = chelsea['Certificate of Habitability Violations - Other Violation']/2
print(chelsea['Certificate of Habitability Violations - Other Violation'].sum())

```

```python
# Collapsing split violations
chelsea['Certificate of Habitability Violations - Use of Lead Paint Prohibited'] = chelsea['Certificate of Habitability Violations - Use of Lead Paint Prohibited'] + chelsea['Certificate of Habitability Violations - Use of Lead Paint Prohibited**']
chelsea.drop(columns=['Certificate of Habitability Violations - Use of Lead Paint Prohibited**'], inplace = True)

chelsea['Certificate of Habitability Violations - Means of Egress'] = chelsea['Certificate of Habitability Violations - Means of Egress'] + chelsea['Certificate of Habitability Violations - Egress Obstruction']
chelsea.drop(columns=['Certificate of Habitability Violations - Egress Obstruction'], inplace = True)
```

### Standardizing size variables

```python
# Standardize variables with log values, ensuring nothing gets turned to -inf
chelsea['log_land_size'] = np.log10(chelsea['Land Size'])
chelsea.log_land_size[chelsea['log_land_size'] < 0] = 0
chelsea['log_building_size'] = np.log10(chelsea['Building Size'])
chelsea.log_building_size[chelsea['log_building_size'] < 0] = 0
chelsea['log_assessed_value'] = np.log10(chelsea['Total Assessed Value'])
chelsea.log_assessed_value[chelsea['log_assessed_value'] < 0] = 0
chelsea['log_bedrooms'] = np.log10(chelsea['Bedrooms'])
chelsea.log_bedrooms[chelsea['log_bedrooms'] < 0] = 0
chelsea['log_bathrooms'] = np.log10(chelsea['Full Bathrooms'])
chelsea.log_bathrooms[chelsea['log_bathrooms'] < 0] = 0
```

### Making string columns more readable

```python
# Making certain columns more readable
chelsea['Current Use'] = chelsea['Current Use'].str[6:]
chelsea['Census Block Group'] = chelsea['Census Block Group'].str[11:]
```

### Labeling predictive variables

```python
# Labeling predictive variable: high risk violations
high_risk = ['Certificate of Habitability Violations - Locks', 'Certificate of Habitability Violations - Smoke Detectors and Carbon Monoxide Alarms', 'Certificate of Habitability Violations - Extermination of Insects, Rodents and Skunks','Certificate of Habitability Violations - Maintenance of Areas Free from Garbage and Rubbish' ]
chelsea['high_risk_violation'] = chelsea[high_risk].sum(axis=1)
chelsea['any_high_risk'] = chelsea.high_risk_violation.mask(chelsea.high_risk_violation > 0, 1)

chelsea.any_high_risk.value_counts()
```

```python
# Labeling predictive variable: any violations
chelsea['violation_total'] = chelsea['Certificate of Habitability Violations']
chelsea.violation_total.fillna(0, inplace = True)
chelsea['any_violation'] = chelsea.violation_total.mask(chelsea.violation_total > 0, 1)

chelsea.any_violation.value_counts()
```

```python
# Labeling predictive variable: overcrowding violations
overcrowding_risk = ['Certificate of Habitability Violations - Locks']

chelsea['overcrowding_risk'] = chelsea[overcrowding_risk].sum(axis=1)
chelsea['any_overcrowding_risk'] = chelsea.overcrowding_risk.mask(chelsea.overcrowding_risk > 0, 1)

chelsea.any_overcrowding_risk.value_counts()
```

```python
# Dropping specific violations
chelsea.drop(chelsea.columns[[range(4, 44)]], axis = 1, inplace = True)
```

### Dummy variables 

```python
# Create dummies for the appropriate variables
use_dummies = pd.get_dummies(chelsea['Current Use'])
census_dummies = pd.get_dummies(chelsea['Census Block Group'])
occupied_dummies = pd.get_dummies(chelsea['Owner Occupied'], prefix = 'owner_occupied_')
fuel_dummies = pd.get_dummies(chelsea['Fuel Type'], prefix = 'fuel_')
heat_dummies = pd.get_dummies(chelsea['Heat Type'], prefix = 'heat_')
opportunity_dummies = pd.get_dummies(chelsea['Opportunity Zones'], prefix='opportunity_zone_')


chelsea = pd.concat([chelsea, use_dummies, census_dummies, occupied_dummies, fuel_dummies, heat_dummies, opportunity_dummies], axis = 1)
```

### Descriptive graphs

```python
# Analyze target area in COH and non-target area in COH
chelsea['analyze'] = chelsea['any_coh'] + chelsea['test parcels']
analyze = chelsea[chelsea['analyze'] != 0]

chelsea_pairplot = analyze[['log_assessed_value','any_violation' , 'Police Incidents After 2015', 'log_building_size', 'analyze', 'Full Bathrooms']]

cols = list(chelsea_pairplot.columns)
cols = [cols[-1]] + cols[:-1]
chelsea_pairplot = chelsea_pairplot[cols]

b= sns.pairplot(chelsea_pairplot, hue = 'analyze', vars=chelsea_pairplot.columns[:-1], diag_kind='kde', palette = 'husl')
```
